```python
demographics = pd.read_csv('./data/unhcr_popstats_export_demographics_all_data.csv',skiprows=3)
                                                                          
                                                                                                
                                                                                    
```

```python
demographics.head()
```

```python
demographics=demographics.convert_objects(convert_numeric=True)
```

```python
ax = demographics.groupby('Year').sum()['F: Total'].plot(kind='area',alpha = 0.9,label= 'Female Total',color='red')
demographics.groupby('Year').sum()['M: Total'].plot(kind ='area',ax=ax,alpha = 0.6,label ='Male Total',color = 'blue')
plt.legend()
plt.show()
```

```python
demographics['Country / territory of asylum/residence'].unique().shape
```

```python
demographics.groupby(['Location Name']).sum()[['F: Total'
                                               ,'M: Total']].sort_values(['F: Total','M: Total']
                                                                                   ,ascending=False)[0:20].plot(kind ='bar',
                                                                                                               title='Location Population')
```

```python
demographics.groupby(['Country / territory of asylum/residence']).sum()[['F: Total',
                                                                         'M: Total']].sort_values(['F: Total',
                                                                                                              'M: Total']
                                                                                                             ,ascending=False)[0:20].plot(kind ='bar',
                                                                                                                                          title='Asylum Population')
```
